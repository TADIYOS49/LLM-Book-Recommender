```python
#import kagglehub
from matplotlib.pyplot import plot_date

import pandas as pd
import warnings
warnings.filterwarnings("ignore")

import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
# if you wanna get the latest data
#path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

#print("Path to dataset files:", path)
```

```python
books = pd.read_csv("books.csv")
books
```

```python
# assigning rows with missing descriptions and changing the age of the book to years
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2024 - books["published_year"]
books.shape[0]
```

```python
# We only take those that have description, num_pages, average_rating, and published_year
book_not_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
book_not_missing.shape[0]
```

```python
book_not_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
book_not_missing.head()
```

```python
# adding the number of words in a description
book_not_missing["words_in_description"] = book_not_missing["description"].str.split().str.len()
```

```python
book_not_missing.head()
```

```python
book_not_missing["words_in_description"].value_counts().reset_index().sort_values("words_in_description",ascending=True).head(25)
```

```python
# Take only descriptions above 25 words for the best result in later searches. 
book_not_missing_25_words = book_not_missing[book_not_missing["words_in_description"] >= 25]
book_not_missing_25_words.shape[0]
```

```python
book_not_missing_25_words
```

```python
book_not_missing_25_words["title_and_subtitle"] = (
    np.where(book_not_missing_25_words["subtitle"].isna(), book_not_missing_25_words["title"],
             book_not_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_not_missing_25_words["tagged_description"] = book_not_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_not_missing_25_words.head()
```

```python
book_not_missing_25_words = book_not_missing_25_words.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
```

```python
book_not_missing_25_words.head()
```

```python
book_not_missing_25_words.to_csv("books_cleaned.csv", index = False)
```
